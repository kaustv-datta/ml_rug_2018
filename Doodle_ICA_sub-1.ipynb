```python
from nilearn.image import smooth_img, index_img, concat_imgs


## Load either run 01, or both run 01 and 02

img = smooth_img('sub-01_task-avmovie_run-1_bold.nii.gz',
                 fwhm=None) 

# img = smooth_img(['sub-01_task-avmovie_run-1_bold.nii.gz',
#                   'sub-01_task-avmovie_run-2_bold.nii.gz'],
#                  fwhm=None) 


## compute mean image (for plotting)
from nilearn import image

mean_img = image.mean_img(img)
```

```python
mean_img.shape
```

```python
from nilearn.decomposition import CanICA

canica = CanICA(n_components=20,
                memory="nilearn_cache", memory_level=2,
                threshold=3., verbose=10, random_state=0)
canica.fit(img)

# Retrieve the independent components in brain space. Directly
# accesible through attribute `components_img_`. Note that this
# attribute is implemented from version 0.4.1. For older versions,
# see note section above for details.
components_img = canica.components_img_
# components_img is a Nifti Image object, and can be saved to a file with
# the following line:
#components_img.to_filename('canica_resting_state.nii.gz')
```

## Results with only the image of Run-01

```python
from nilearn.plotting import plot_prob_atlas

# Plot all ICA components together
plot_prob_atlas(components_img, bg_img = mean_img,display_mode='z', title='All ICA components')
```

```python
from nilearn.image import iter_img
from nilearn.plotting import plot_stat_map, show

for i, cur_img in enumerate(iter_img(components_img)):
    plot_stat_map(cur_img, bg_img=mean_img, display_mode="ortho", title="IC %d" % i,
                  colorbar=True,draw_cross=False)

show()
```

## Results with images of Run-01 and Run-02. Both analyses seem to give similar outputs...
